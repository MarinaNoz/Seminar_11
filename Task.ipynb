```python
from sympy import *
```

```python
x = Symbol('x',real = True)
```

```python
func = x**2 - 6*abs(x) + 8
func
```

* Определить корни

```python
roots = solve(func)
print(f'корни', roots)
```

* Построить график

```python
graf_y = plot(func)
```

вычислим производную функции

```python
derivate = diff(func)
derivate
```

```python
function = plot(derivate)
```

Излом функции в точке х = 0. Производная равна 0 в точках -3, 3. Добавим точки в корни  производной:

```python
point = solve(derivate)
```

```python
point.append(point[0] * - 1)
point.append(0)
point = sorted(point)
point
```

* Вычислим вершины

```python
x_1 = -3
x_2 = 3
for i in point:
    x_1 = i-0.01 if i == 0 else i-0.01*abs(i)
    x_2 = i+0.01 if i == 0 else i+0.01*abs(i)
    if derivate.evalf(subs={x: x_1}) < derivate.evalf(subs={x: i}) and derivate.evalf(subs={x: x_2}) > derivate.evalf(subs={x: i}):
        print(f'точка минимума - "{i}"')
    else:
        print(f'точка максимума - "{i}"')
```

* Найти интервалы, на которых функция возрастает и убывает

```python
def interval_func(x):
    interval = []
    interval.append([x[0]-1,x[0]])
    for i in range(len(x)-1):
        r = [x[i], x[i+1]]
        interval.append(r)
    interval.append([x[-1], x[-1]+1])
    return interval

intervals = interval_func(point)
for i in intervals:
    start = -oo if intervals.index(i) == 0 else i [0]
    end = oo if intervals.index(i) == len(intervals)-1 else i[1]
    if derivate.evalf(subs = {x:(i[0] + i[1])/ 2}) > 0:
        print(f'функция возрастает на интервале {start, end}')
    else:
        print(f'функция убывает на интервале {start, end}')
```
